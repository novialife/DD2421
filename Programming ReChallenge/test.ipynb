```python
import pandas as pd
mervans = open("predictions.txt", "r")
daniels = open("Classes.txt", "r")

df_merv = pd.Series(mervans.read().splitlines())
df_dan = pd.Series(daniels.read().splitlines())
```

```python
count = 0
for i in range(len(df_merv)):
    if df_merv[i] == df_dan[i]:
        count += 1

print(count)
print(len(df_merv) - count)
```

```python
df_merv.value_counts()
```
